```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D
from itertools import combinations
file_path = '/content/properties segregation (Autosaved).xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')
print("Shape of the DataFrame:", df.shape)
print("Column names in the DataFrame:", df.columns)
print("First few rows of the DataFrame:")
df = pd.read_excel(file_path)
df.rename(columns=lambda x: x.strip(), inplace=True)
print(df.head())
print("Null values in the DataFrame:")
total_compounds = df['Compound Name'].count()
print("Total number of compounds:", total_compounds)
print(df.isnull().sum())
print("Total compounds:", len(df))
print("Columns in the DataFrame (similar to data.keys()):", df.columns)
df.head()
required_columns = ["Anti-Inflammatory", "Anti-Oxidant", "Anti-Cancer", "X properties"]
for col in required_columns:
    if col not in df.columns:
        print(f"Warning: Column '{col}' is missing from the DataFrame.")
        df[col] = 0
        if 'Cluster' not in df.columns:
          feature_columns = ["Anti-Inflammatory", "Anti-Oxidant", "Anti-Cancer", "X properties"]
          scaler = StandardScaler()
          scaled_features = scaler.fit_transform(df[feature_columns])
    n_clusters = 5
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    features = ["Anti-Inflammatory", "Anti-Oxidant", "Anti-Cancer", "X properties"]
df['Cluster'] = kmeans.fit_predict(df[features])
print("Number of compounds in each cluster:")
print(df.head())
print(df['Cluster'].value_counts())
cluster_counts = df['Cluster'].value_counts()
cluster_means = df.groupby('Cluster')[['Anti-Inflammatory', 'Anti-Oxidant', 'Anti-Cancer', 'X Properties']].mean()
print("Mean Values for Each Cluster:")
print(cluster_means)
import matplotlib.pyplot as plt
ax = cluster_means.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Values of Properties by Cluster')
plt.ylabel('Mean Value')
plt.xlabel('Cluster')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.legend(title='Properties')
for i, count in enumerate(cluster_counts):
   if i < len(ax.patches):
    bar_height = ax.patches[i].get_height()  # Get height of the i-th bar
    ax.text(i, bar_height + 0.02, str(count), ha='center', va='bottom', fontsize=12)
plt.show()
fig = plt.figure(figsize=(6, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(
    df['Anti-Oxidant'],
    df['Anti-Inflammatory'],
    df['Anti-Cancer'],
    c=df['Cluster'],
    s=df['X Properties'] * 100,
    cmap='Set1',
    alpha=0.6,
    edgecolor='k'
)
ax.set_title('3D Scatter Plot of K-Means Clustering of Compounds')
ax.set_xlabel('Anti-Oxidant')
ax.set_ylabel('Anti-Inflammatory')
ax.set_zlabel('Anti-Cancer')
legend = ax.legend(*scatter.legend_elements(), title='Clusters', loc='upper right')
ax.add_artist(legend)
cluster_counts = df['Cluster'].value_counts()
for cluster in cluster_counts.index:
    count = cluster_counts[cluster]
    ax.text(
        x=df[df['Cluster'] == cluster]['Anti-Oxidant'].mean(),
        y=df[df['Cluster'] == cluster]['Anti-Inflammatory'].mean(),
        z=df[df['Cluster'] == cluster]['Anti-Cancer'].mean(),
        s=f'N={count}',
        color='black',
        fontsize=10,
        weight='bold'
    )
    plt.show()
    cluster_compounds = {}
    for cluster in df['Cluster'].unique():
      cluster_data = df[df['Cluster'] == cluster]
      cluster_compounds[cluster] = cluster_data
      print(f"\nCluster {cluster}:")
    print(cluster_data[['Compound Name', 'Anti-Oxidant', 'Anti-Inflammatory', 'Anti-Cancer', 'X Properties']])
cluster_distribution = df['Cluster'].value_counts().reset_index()
cluster_distribution.columns = ['Cluster', 'Number of Compounds']
print(cluster_distribution)
cluster_distribution.to_excel('Cluster_Distribution.xlsx', index=False)
grouped = df.groupby('Cluster')['Compound Name'].apply(list).reset_index()
grouped.columns = ['Cluster', 'Compounds']
compounds_expanded = grouped['Compounds'].apply(pd.Series)
compounds_expanded.columns = [f'Compound {i+1}' for i in range(compounds_expanded.shape[1])]
final_output = pd.concat([grouped[['Cluster']], compounds_expanded], axis=1)
final_output.to_excel('Compounds_by_Cluster.xlsx', index=False)
print(final_output)

plt.scatter(df[df['Cluster'] == 0]['Anti-Oxidant'], df[df['Cluster'] == 0]['Anti-Inflammatory'], s=100, c='red', label='Cluster 0')
plt.scatter(df[df['Cluster'] == 1]['Anti-Oxidant'], df[df['Cluster'] == 1]['Anti-Inflammatory'], s=100, c='blue', label='Cluster 1')
plt.scatter(df[df['Cluster'] == 2]['Anti-Oxidant'], df[df['Cluster'] == 2]['Anti-Inflammatory'], s=100, c='green', label='Cluster 2')
plt.scatter(df[df['Cluster'] == 3]['Anti-Oxidant'], df[df['Cluster'] == 3]['Anti-Inflammatory'], s=100, c='orange', label='Cluster 3')
plt.title('Scatter Plot of Clusters Based on Anti-Oxidant and Anti-Inflammatory Properties')
plt.xlabel('Anti-Oxidant')
plt.ylabel('Anti-Inflammatory')
plt.legend(title='Clusters')
plt.grid(True)

# Show plot
plt.show()
```
